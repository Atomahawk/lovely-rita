Functions

```python
def get_data(data_files, dtype, column_map, delimiter):
    df = pd.concat([pd.read_csv(f, dtype=dtype, delimiter=delimiter) for f in data_files]).reset_index(drop=True)
    df.rename(columns=column_map, inplace=True)
    return df

def get_column_report(df):
    column_report = []
    for column in df.columns:
        unique = df[column].unique()
        sample = np.nan
        for value in unique:
            if value is not np.nan:
                sample = value
                break
        nans = df[column].isnull().sum()
        pct_nan = 100. * nans / df.shape[0]
        column_report.append([column, df[column].dtype, len(unique), sample, nans, pct_nan])
    columns=["Column Name", "Data Type", "Unique Count", "Sample Value", "NaNs", "% NaN"]
    column_report = pd.DataFrame(column_report, columns=columns).round(2)
    column_report.sort_values(by="NaNs", inplace=True)
    return column_report

def get_uniques(df):
    for column in df.columns:
        print column, df[column].unique()
        
def output_addresses(condition, file_out):
    addresses = set()
    for i, item in (df[condition]["Street"] + " " + df[condition]["City"] + " " + df[condition]["State"]).iteritems():
        addresses.add(" ".join(item.lower().split()))
    with open(file_out, 'w') as output:
        for address in addresses:
            output.write(address + '\n')
    return addresses

def convert_time_str_to_min(time_str):
    time_elements = time_str.split(":")
    hours = int(time_elements[0])
    minutes = int(time_elements[1])
    return minutes + (hours * 60)

def convert_min_to_time_str(minutes):
    minutes_str = str(minutes % 60)
    if len(minutes_str) == 1:
        minutes_str = '0' + minutes_str
    return str((minutes / 60) % 24) + ":" + minutes_str
    
def get_time_diff(time_start, time_end):
    time_1 = convert_time_str_to_min(time_start)
    time_2 = convert_time_str_to_min(time_end)
    if time_2 < time_1:
        time_2 += 24 * 60
    return time_2 - time_1
```

There are 0 missing dates and 119 missing times.
Since the data is recorded in temporal order, the times in the neighboring rows represent a window in which the missing time exists.  We can impute the missing time values as the mean time of the neighboring rows.  There are 8 cases where missing values are next to eachother.  For those, we can find the next existing value, and divide the time evenly to fill in the missing values.

```python
for i in np.where(df["Ticket Issue Time"].isnull())[0]:
    prev_value = np.nan
    prev_index = i
    while type(prev_value) != str:
        prev_index -= 1
        prev_value = df.iloc[prev_index]["Ticket Issue Time"]
    #print type(prev_value)
    #print prev_value
    next_value = np.nan
    next_index = i
    while type(next_value) != str:
        next_index += 1
        next_value = df.iloc[next_index]["Ticket Issue Time"]
    diff = get_time_diff(prev_value, next_value)
    split_count = next_index-prev_index
    prev_minutes = convert_time_str_to_min(prev_value)
    for _ in range(split_count):
        prev_minutes += (diff / split_count)
        df.at[i, "Ticket Issue Time"] = convert_min_to_time_str(prev_minutes)
```
